```python
from PIL import Image, ImageDraw

# # --- Configuration List ---
# # We put all the image paths and box data into a single list to loop through.
# all_images_data = [
#     {
#         "image_path": "image_01.jpg",
#         "output_path": "image_01_with_boxes.jpg",
#         "boxes_data": [
#             {"box_2d": [257, 362, 456, 448], "label": "Tunnel 1"},
#             {"box_2d": [444, 469, 563, 622], "label": "Tunnel 2"},
#         ],
#     },
#     {
#         "image_path": "image_02.jpg",
#         "output_path": "image_02_with_boxes.jpg",
#         "boxes_data": [
#             {
#                 "box_2d": [337, 187, 868, 309],
#                 "label": "Main tunnel running along the retaining wall edge",
#             },
#             {
#                 "box_2d": [146, 362, 350, 566],
#                 "label": "Cross-lawn tunnel connecting edge activity to the center",
#             },
#         ],
#     },
#     {
#         "image_path": "image_03.jpg",
#         "output_path": "image_03_with_boxes.jpg",
#         "boxes_data": [
#             {
#                 "box_2d": [258, 161, 800, 290],
#                 "label": "Main tunnel running along the retaining wall edge",
#             },
#             {
#                 "box_2d": [178, 477, 342, 750],
#                 "label": "Main tunnel routing from the center toward the driveway edge",
#             },
#         ],
#     },
#     {
#         "image_path": "image_04.jpg",
#         "output_path": "image_04_with_boxes.jpg",
#         "boxes_data": [
#             {
#                 "box_2d": [280, 820, 750, 980],
#                 "label": "Main tunnel running along the concrete wall boundary",
#             },
#             {
#                 "box_2d": [420, 480, 600, 720],
#                 "label": "Tunnel connecting the fresh center mound to the older dead patch",
#             },
#         ],
#     },
#     {
#         "image_path": "image_04.jpg",
#         "output_path": "rated_1.jpg",
#         "boxes_data": [
#             {
#                 "box_2d": [280, 820, 750, 980],
#                 "label": "Main tunnel running along the concrete wall boundary",
#             },
#             #         {
#             #             "box_2d": [420, 480, 600, 720],
#             #             "label": "Tunnel connecting the fresh center mound to the older dead patch",
#             #         },
#         ],
#     },
#     {
#         "image_path": "image_02.jpg",
#         "output_path": "rated_2.jpg",
#         "boxes_data": [
#             {
#                 "box_2d": [337, 187, 868, 309],
#                 "label": "Main tunnel running along the retaining wall edge",
#             },
#             # {
#             #     "box_2d": [146, 362, 350, 566],
#             #     "label": "Cross-lawn tunnel connecting edge activity to the center",
#             # },
#         ],
#     },
#     {
#         "image_path": "image_01.jpg",
#         "output_path": "rated_3.jpg",
#         "boxes_data": [
#             {"box_2d": [257, 362, 456, 448], "label": "Tunnel 1"},
#             # {"box_2d": [444, 469, 563, 622], "label": "Tunnel 2"},
#         ],
#     },
# ]
# # ---------------------

# --- Configuration List for Images 5, 6, and 7 ---
all_images_data = [
    {
        "image_path": "image_05.jpg",
        "output_path": "image_05_with_boxes.jpg",
        "boxes_data": [
            {
                "box_2d": [480, 350, 750, 480],
                "label": "Likely main runway parallel to the brick border connecting the two mounds",
            }
        ],
    },
    {
        "image_path": "image_06.jpg",
        "output_path": "image_06_with_boxes.jpg",
        "boxes_data": [
            {
                "box_2d": [420, 200, 550, 420],
                "label": "Tunnel connecting the fresh center mound to the older patch on the left",
            }
        ],
    },
    {
        "image_path": "image_07.jpg",
        "output_path": "image_07_with_boxes.jpg",
        "boxes_data": [
            {
                "box_2d": [520, 400, 680, 650],
                "label": "Horizontal tunnel connecting fresh mound to the older patches on the left",
            },
            {
                "box_2d": [680, 500, 750, 900],
                "label": "Potential travel path riding right along the gravel border edge",
            },
        ],
    },
]
# ---------------------


def draw_boxes(img_path, out_path, boxes):
    try:
        img = Image.open(img_path)
        draw = ImageDraw.Draw(img)
        width, height = img.size

        for item in boxes:
            box = item["box_2d"]
            # Convert 0-1000 scale to actual image pixels
            ymin, xmin, ymax, xmax = box
            left = (xmin / 1000.0) * width
            top = (ymin / 1000.0) * height
            right = (xmax / 1000.0) * width
            bottom = (ymax / 1000.0) * height

            # Draw the rectangle (Red outline, width of 5 pixels)
            draw.rectangle([left, top, right, bottom], outline="red", width=5)

        img.save(out_path)
        print(f"Success! Saved image with boxes to {out_path}")

    except FileNotFoundError:
        print(
            f"Error: Could not find the file {img_path}. Make sure the image is in the same folder as this script."
        )
    except Exception as e:
        print(f"An error occurred while processing {img_path}: {e}")


# --- Main Loop ---
# This goes through the list above and processes each image one by one.
for data in all_images_data:
    draw_boxes(data["image_path"], data["output_path"], data["boxes_data"])
```
