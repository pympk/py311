```python
import os # Only needed if you have full paths, but good practice

def get_unique_dates_from_filenames(filename_list):
  """
  Extracts unique date strings (YYYY-MM-DD) from the beginning of filenames.

  Args:
    filename_list: A list of strings, where each string is expected
                   to start with a 'YYYY-MM-DD' date format.

  Returns:
    A sorted list of unique date strings found in the filenames.
    Returns an empty list if the input list is empty.
  """
  if not filename_list:
    return []

  # Use a set comprehension to extract the first 10 chars and ensure uniqueness
  # os.path.basename is used in case the list contains full paths
  unique_dates_set = {os.path.basename(fname)[:10] for fname in filename_list}

  # Convert the set to a list and sort it (sets are unordered)
  return sorted(list(unique_dates_set))
```
