```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns

def create_rank_history_df(file_list, data_dir):
    """
    Reads daily files to compile a history of ranks for each ticker.

    Args:
        file_list (list): A list of sorted parquet filenames.
        data_dir (Path): The directory where files are stored.

    Returns:
        pd.DataFrame: A DataFrame with tickers as the index, dates as columns,
                      and ranks as the values. NaN indicates the ticker was not
                      present on that day.
    """
    daily_ranks_list = []
    for filename in file_list:
        # Extract the date string from the start of the filename
        date_str = filename.split('_')[0]
        
        # Read the daily data file
        df_daily = pd.read_parquet(data_dir / filename)
        
        # Extract the 'Rank' column (it's a Series with tickers as its index)
        ranks_series = df_daily['Rank']
        
        # Rename the Series to the date, which will become the column name
        ranks_series.name = pd.to_datetime(date_str)
        
        daily_ranks_list.append(ranks_series)

    # Concatenate all Series into a single DataFrame, aligning on the ticker index
    df_rank_history = pd.concat(daily_ranks_list, axis=1)
    
    # Sort columns by date just in case the file list wasn't perfectly sorted
    df_rank_history = df_rank_history.sort_index(axis=1)
    
    return df_rank_history

# --- Example Usage ---
# Assuming DATA_DIR is a pathlib.Path to your data folder
# and file_list is your list of filenames.
# DATA_DIR = Path('your_data_directory/')
df_rank_history = create_rank_history_df(file_list, DATA_DIR)

print("Shape of the resulting DataFrame:", df_rank_history.shape)
print(df_rank_history.head(30))
```

```python
import pandas as pd
import numpy as np

def find_dip_opportunities(df_rank_history, lookback_days=60, recent_days=10, 
                           min_trend_improvement=10, min_recent_drop=5,
                           current_rank_start=1, current_rank_end=None):
    """
    Scans rank history to find tickers that had a steady uptrend in rank
    followed by a recent significant drop, within a specified current rank range.

    Args:
        df_rank_history (pd.DataFrame): DataFrame with tickers as index, dates as columns, and ranks as values.
        lookback_days (int): The number of days to define the "past performance" period.
        recent_days (int): The number of recent days to check for a "dip".
        min_trend_improvement (int): The minimum number of ranks a ticker must have improved
                                     during the lookback period to be considered.
        min_recent_drop (int): The minimum number of ranks a ticker must have dropped
                               in the recent period to be considered a "dip".
        current_rank_start (int): The minimum current rank to be included in the analysis. Defaults to 1.
        current_rank_end (int, optional): The maximum current rank to be included. If None, no upper limit.

    Returns:
        list: A sorted list of dictionaries, where each dictionary contains information
              about a potential "buy the dip" opportunity.
    """
    
    # --- Guard Clause to check for sufficient data ---
    total_days_needed = lookback_days + recent_days
    available_days = len(df_rank_history.columns)
    
    if available_days < total_days_needed:
        print(f"Error: Not enough data for analysis. Need {total_days_needed} (lookback_days {lookback_days} + recent_days {recent_days}) days, but only {available_days} are available.")
        return []
        
    # --- Date Setup ---
    all_dates = df_rank_history.columns
    last_date = all_dates[-1]
    recent_period_start_date = all_dates[-recent_days]
    lookback_period_end_date = all_dates[-(recent_days + 1)]
    lookback_period_start_date = all_dates[-(recent_days + lookback_days)]
    
    lookback_dates = df_rank_history.loc[:, lookback_period_start_date:lookback_period_end_date].columns
    recent_dates = df_rank_history.loc[:, recent_period_start_date:last_date].columns

    opportunities = []

    # --- NEW: Pre-filter tickers based on their current rank ---
    # This significantly speeds up the loop by reducing the number of tickers to check.
    ranks_on_last_day = df_rank_history[last_date]
    
    # Create boolean masks for the start and end ranks
    mask_start = (ranks_on_last_day >= current_rank_start)
    if current_rank_end is not None:
        mask_end = (ranks_on_last_day <= current_rank_end)
        tickers_to_scan = df_rank_history[mask_start & mask_end].index
    else: # If no end rank is specified, only use the start mask
        tickers_to_scan = df_rank_history[mask_start].index
        
    print(f"Analyzing {len(tickers_to_scan)} tickers currently ranked between {current_rank_start} and {current_rank_end or 'max'}...")

    # --- Loop through the PRE-FILTERED list of tickers ---
    for ticker in tickers_to_scan:
        lookback_ranks = df_rank_history.loc[ticker, lookback_dates].dropna()
        recent_ranks = df_rank_history.loc[ticker, recent_dates].dropna()

        # (The rest of the logic inside the loop is unchanged)
        if len(lookback_ranks) < 5 or len(recent_ranks) < 3:
            continue

        rank_at_start_of_lookback = lookback_ranks.iloc[0]
        rank_at_end_of_lookback = lookback_ranks.iloc[-1]
        trend_improvement = rank_at_start_of_lookback - rank_at_end_of_lookback
        
        if trend_improvement < min_trend_improvement:
            continue
            
        x = np.arange(len(lookback_ranks))
        slope, _ = np.polyfit(x, lookback_ranks, 1)
        
        if slope >= 0:
            continue

        rank_at_start_of_recent = recent_ranks.iloc[0]
        rank_at_end_of_recent = recent_ranks.iloc[-1]
        recent_drop = rank_at_end_of_recent - rank_at_start_of_recent

        if recent_drop < min_recent_drop:
            continue
        
        opportunities.append({
            'ticker': ticker,
            'current_rank': int(rank_at_end_of_recent),
            'previous_rank': int(rank_at_start_of_recent),
            'recent_drop': int(recent_drop),
            'lookback_slope': round(slope, 2)
        })

    sorted_opportunities = sorted(opportunities, key=lambda x: x['recent_drop'], reverse=True)
    
    return sorted_opportunities

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mplcursors
from adjustText import adjust_text

def plot_rank_changes_for_tickers(df_rank_history, ticker_list, title_suffix="", 
                                  exclude_tickers=None, figure_height=10, figure_width=12):
    """
    Plots rank history with a layout optimized to fit within a standard
    notebook output frame without horizontal scrolling.

    Args:
        df_rank_history (pd.DataFrame): The full rank history data.
        ticker_list (list): The specific tickers to plot.
        title_suffix (str, optional): Text to add to the plot title.
        exclude_tickers (list, optional): A list of tickers to manually remove.
        figure_height (int, optional): The exact height of the plot in inches. Defaults to 10.
        figure_width (int, optional): The exact width of the plot in inches. Defaults to 12.
    """
    
    plt.close('all')

    # --- Data Filtering ---
    df_to_plot = df_rank_history.loc[ticker_list]
    
    if exclude_tickers:
        tickers_to_remove = [t for t in exclude_tickers if t in df_to_plot.index]
        if tickers_to_remove:
            print(f"Manually excluding {len(tickers_to_remove)} tickers: {tickers_to_remove}")
            df_to_plot = df_to_plot.drop(labels=tickers_to_remove)

    num_tickers = len(df_to_plot)
    if num_tickers == 0:
        print("No tickers remaining to plot after filtering.")
        return
        
    sns.set_style("whitegrid")
    fig, ax = plt.subplots(figsize=(figure_width, figure_height))
    
    df_to_plot.T.plot(ax=ax, marker='.', markersize=4, legend=False)
    
    # --- End-of-Line Labels ---
    texts = []
    for ticker, data in df_to_plot.iterrows():
        last_valid_idx = data.last_valid_index()
        if last_valid_idx is not None:
            last_rank = data[last_valid_idx]
            texts.append(ax.text(last_valid_idx, last_rank, ticker, fontsize=9))

    adjust_text(
        texts, ax=ax, 
        force_text=(0.1, 0.25),
        arrowprops=dict(arrowstyle='-', color='gray', lw=0.5)
    )

    # --- Interactive Hovering ---
    cursor = mplcursors.cursor(ax.lines, hover=True)
    @cursor.connect("add")
    def on_add(sel):
        ticker_name = sel.artist.get_label()
        rank_value = sel.target[1]
        sel.annotation.set_text(f'Ticker: {ticker_name}\nRank: {rank_value:.0f}')
        sel.annotation.get_bbox_patch().set(facecolor='lightblue', alpha=0.7)
        sel.annotation.arrow_patch.set(arrowstyle="->", facecolor="black", alpha=0.5)

    # --- Final Touches ---
    ax.invert_yaxis()
    ax.set_title(f'Rank Movement for {title_suffix}' if title_suffix else 'Rank Movement for Selected Tickers', fontsize=16)
    ax.set_xlabel('Date', fontsize=12)
    ax.set_ylabel('Rank (Lower is Better)', fontsize=12)
    
    # --- Margin Control (unchanged, but now works with the smaller figure size) ---
    fig.subplots_adjust(left=0.07, right=0.85, top=0.95, bottom=0.1)

    ###################
    # Extend X-Axis to the right to make more room for ticker labels
    # Get the first and last dates from the data being plotted
    start_date = df_to_plot.columns.min()
    last_date = df_to_plot.columns.max()

    # Calculate the total duration of the plotted data
    duration = last_date - start_date

    # Extend the x-axis by 8% of the total duration to create space for labels
    padding = duration * 0.08
    ax.set_xlim(right=last_date + padding)
    ###################

    plt.show()
    
```

```python
import pandas as pd
import numpy as np

def find_reversal_opportunities(df_rank_history, lookback_days=60, recent_days=10, 
                                min_trend_improvement=10, min_dip_magnitude=10, min_reversal_improvement=5,
                                current_rank_start=1, current_rank_end=None):
    """
    Finds tickers with a strong past trend that recently dipped and are now showing signs of reversal.

    Args:
        df_rank_history (pd.DataFrame): DataFrame of rank history.
        lookback_days (int): Period to establish the long-term uptrend.
        recent_days (int): Period to identify the dip-and-reversal pattern.
        min_trend_improvement (int): Minimum rank improvement during the lookback period.
        min_dip_magnitude (int): The dip must be at least this many ranks worse than the start of the recent period.
        min_reversal_improvement (int): The current rank must be at least this many ranks better than the bottom of the dip.
        current_rank_start (int): The minimum current rank to be included in the analysis.
        current_rank_end (int, optional): The maximum current rank to be included.

    Returns:
        list: A sorted list of dictionaries representing reversal opportunities.
    """
    
    # --- Guard Clause and Date Setup (same as before) ---
    total_days_needed = lookback_days + recent_days
    available_days = len(df_rank_history.columns)
    
    if available_days < total_days_needed:
        print(f"Error: Not enough data. Need {total_days_needed} days, but only {available_days} available.")
        return []
        
    all_dates = df_rank_history.columns
    last_date = all_dates[-1]
    recent_period_start_date = all_dates[-recent_days]
    lookback_period_end_date = all_dates[-(recent_days + 1)]
    lookback_period_start_date = all_dates[-(recent_days + lookback_days)]
    
    lookback_dates = df_rank_history.loc[:, lookback_period_start_date:lookback_period_end_date].columns
    recent_dates = df_rank_history.loc[:, recent_period_start_date:last_date].columns

    opportunities = []

    # --- Pre-filter tickers (same as before) ---
    ranks_on_last_day = df_rank_history[last_date]
    mask_start = (ranks_on_last_day >= current_rank_start)
    if current_rank_end is not None:
        mask_end = (ranks_on_last_day <= current_rank_end)
        tickers_to_scan = df_rank_history[mask_start & mask_end].index
    else:
        tickers_to_scan = df_rank_history[mask_start].index
        
    print(f"Analyzing {len(tickers_to_scan)} tickers for reversal patterns...")

    # --- Loop through the pre-filtered tickers ---
    for ticker in tickers_to_scan:
        lookback_ranks = df_rank_history.loc[ticker, lookback_dates].dropna()
        recent_ranks = df_rank_history.loc[ticker, recent_dates].dropna()

        if len(lookback_ranks) < 5 or len(recent_ranks) < 3:
            continue

        # --- Criterion 1: Strong Past Performance (Lookback Period) ---
        # This logic is unchanged.
        if (lookback_ranks.iloc[0] - lookback_ranks.iloc[-1]) < min_trend_improvement:
            continue
        x = np.arange(len(lookback_ranks))
        slope, _ = np.polyfit(x, lookback_ranks, 1)
        if slope >= 0:
            continue

        # --- Criterion 2: Recent Dip and Reversal Pattern ---
        # This is the new core logic.
        rank_at_start_of_recent = recent_ranks.iloc[0]
        current_rank = recent_ranks.iloc[-1]
        dip_rank = recent_ranks.max() # Find the absolute worst rank (the bottom of the dip)

        # a) Was the dip significant?
        dip_magnitude = dip_rank - rank_at_start_of_recent
        if dip_magnitude < min_dip_magnitude:
            continue
        
        # b) Has it started to reverse from the bottom of the dip?
        reversal_improvement = dip_rank - current_rank
        if reversal_improvement < min_reversal_improvement:
            continue
        
        # If all criteria are met, it's a valid opportunity
        opportunities.append({
            'ticker': ticker,
            'current_rank': int(current_rank),
            'dip_rank': int(dip_rank),
            'reversal_improvement': int(reversal_improvement),
            'lookback_slope': round(slope, 2)
        })

    # Sort results by the strongest reversal improvement
    sorted_opportunities = sorted(opportunities, key=lambda x: x['reversal_improvement'], reverse=True)
    
    return sorted_opportunities
```
