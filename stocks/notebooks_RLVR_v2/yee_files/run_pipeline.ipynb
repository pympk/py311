```python
import os
import subprocess

# Install papermill if it's not already installed
!pip install papermill

lookbacks = [10, 20]
os.environ["CACHE_START_DATE"] = "2026-01-01"

# 2. Define your notebooks in sequence
# notebooks = ["0_prepare_data.ipynb", "1_train_model.ipynb", "2_evaluate.ipynb"]
notebooks = [
    "/content/drive/MyDrive/Colab Notebooks/RLVR_Part1_AlphaCache_v2.ipynb",
]

for lookback in lookbacks:
    print(f"\n{'#' * 50}")
    print(f"### STARTING PIPELINE WITH CACHE_LOOKBACK = {lookback} ###")
    print(f"{ '#' * 50}\n")

    os.environ["CACHE_LOOKBACK"] = str(lookback)

    for nb in notebooks:
        print(f"\n========== Running {nb} ==========")

        # Extract directory and filename for output notebook name
        input_dir = os.path.dirname(nb)
        nb_filename = os.path.basename(nb)
        out_nb_filename = nb_filename.replace(".ipynb", f"_lookback_{lookback}.ipynb")
        out_nb = os.path.join(input_dir, out_nb_filename)

        # Papermill command:
        # --log-output streams your print() statements and tqdm bars to the terminal!
        cmd = [
            "papermill",
            nb,  # Input notebook
            out_nb,  # Output notebook
            "--log-output",
        ]

        subprocess.run(cmd, check=True)

print("\nAll pipeline runs completed successfully!")
```
