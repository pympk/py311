```python
import pandas_datareader.data as web
import datetime
import pandas as pd

# Set the start date to 1970 to ensure we get the full series history
start_date = datetime.datetime(1970, 1, 1)
end_date = datetime.datetime.now()

tickers = ["BAMLH0A0HYM2", "T10Y2Y"]

try:
    # Fetch all available data from FRED
    df = web.DataReader(tickers, "fred", start_date, end_date)
    df.columns = ["High_Yield_Spread", "Yield_Curve_10Y2Y"]

    # Drop rows where BOTH are missing (to keep the early Yield Curve data)
    df_clean = df.dropna(how="all")

    print(f"Data successfully downloaded.")
    print(
        f"Yield Curve (T10Y2Y) starts: {df['Yield_Curve_10Y2Y'].first_valid_index().date()}"
    )
    print(
        f"High Yield Spread (BAMLH0A0HYM2) starts: {df['High_Yield_Spread'].first_valid_index().date()}"
    )

    # Save the full history to a CSV
    df_clean.to_csv("output/High_Yield_Spread_T10Y2Y_Spread.csv")
    df_clean.to_parquet(
        "output/High_Yield_Spread_T10Y2Y_Spread.parquet", engine="pyarrow"
    )
    print("\nHigh Yield Spread saved to 'output/High_Yield_Spread_T10Y2Y_Spread.csv'")

except Exception as e:
    print(f"Error fetching data: {e}")
```
